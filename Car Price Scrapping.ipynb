```python
# defining variable
model_num=[]
brand=[]
name=[]
variant=[]
distance=[]
owner=[]
fuel_type=[]
price=[]

#extracting model_num
try:
    model_tags=driver.find_elements_by_xpath("//div[@class='_1jpRU']/h2")
    for i in model_tags:
        models=i.text.split(' ')[0]
        model_num.append(models)
    model_num[0:]
except NoSuchElementException:
    model_num.append('-')
    
#extracting car brand
try:
    brand_tags=driver.find_elements_by_xpath("//div[@class='_1jpRU']/h2")
    #battery=[]
    for i in brand_tags:
        brands=i.text.split(' ')[1:2]
        b="[]".join(brands)
        brand.append(b)
    brand[0:]
except NoSuchElementException:
    brand.append('-')
time.sleep(1)

#extracting car name
try:
    name_tags=driver.find_elements_by_xpath("//div[@class='_1jpRU']/h2")
    for i in name_tags:
        names=i.text.split(' ')[2:]
        l=" ".join(names)
        name.append(l)
    name[0:]
except NoSuchElementException:
    name.append('-')
time.sleep(1)

#extracting car variant
try:
    variant_tags=driver.find_elements_by_xpath("//div[@class='_1jpRU']/p/span")
    for i in variant_tags:
        variants=i.text
        variant.append(variants)
    variant[0:]
except NoSuchElementException:
    variant.append('-')
time.sleep(1)
    
#extracting distance travelled
try:
    distance_tags=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[1]")
    for i in distance_tags:
        distances=i.text.split(' ')[0]
        d=distances.replace(',','')
        #d=" ".join(distances)
        distance.append(d)
    distance[0:]
except NoSuchElementException:
    distance.append('-')
time.sleep(1)
    
#extracting number of owners
try:
    owner_tags=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[2]")
    for i in owner_tags:
        owners=i.text.split(' ')[0]
        owner.append(owners)
    owner[0:]
except NoSuchElementException:
    owner.append('-')
    
#extracting fuel type
try:
    fuel_tags=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[3]")
    for i in fuel_tags:
        #fuels=i.text
        fuel_type.append(i.text)
    fuel_type[0:]
except NoSuchElementException:
    fuel_type.append('-')
time.sleep(1)
    
#extracting price
try:
    price_tags=driver.find_elements_by_xpath("//div[@class='_7udZZ']/span")
    for i in price_tags:
        prices=i.text.replace('₹','')
        p=prices.replace(',','')
        price.append(p)
    price[0:]
except NoSuchElementException:
    price.append('-')
time.sleep(1)


NDel1=pd.DataFrame({'Model Year': model_num[0:640],'Brand': brand[0:640],
                 'Car Name': name[0:640],'Variant': variant[0:640],'Distance Travelled': distance[0:640],
                 'Number of Owners': owner[0:640],'fuel type': fuel_type[0:640],'Price': price[0:640]})
```

```python
# defining variable
model_num=[]
brand=[]
name=[]
variant=[]
distance=[]
owner=[]
fuel_type=[]
price=[]

#extracting model_num
try:
    model_tags=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
    for i in model_tags:
        models=i.text.split(' ')[0]
        model_num.append(models)
    model_num[0:]
except NoSuchElementException:
    model_num.append('-')
    
#extracting car brand
try:
    brand_tags=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
    #battery=[]
    for i in brand_tags:
        brands=i.text.split(' ')[1:2]
        b="[]".join(brands)
        brand.append(b)
    brand[0:]
except NoSuchElementException:
    brand.append('-')
time.sleep(1)

#extracting car name
try:
    name_tags=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
    for i in name_tags:
        names=i.text.split(' ')[2:]
        l=" ".join(names)
        name.append(l)
    name[0:]
except NoSuchElementException:
    name.append('-')
time.sleep(1)

#extracting car variant
try:
    variant_tags=driver.find_elements_by_xpath("//div[@class='truncate dotlist']/span[3]")
    for i in variant_tags:
        variants=i.text
        variant.append(variants)
    variant[0:]
except NoSuchElementException:
    variant.append('-')
time.sleep(1)
    
#extracting distance travelled
try:
    distance_tags=driver.find_elements_by_xpath("//div[@class='truncate dotlist']/span[1]")
    for i in distance_tags:
        distances=i.text.split(' ')[0]
        d=distances.replace(',','')
        #d=" ".join(distances)
        distance.append(d)
    distance[0:]
except NoSuchElementException:
    distance.append('-')
time.sleep(1)
    

#extracting fuel type
try:
    fuel_tags=driver.find_elements_by_xpath("//div[@class='truncate dotlist']/span[2]")
    for i in fuel_tags:
        #fuels=i.text
        fuel_type.append(i.text)
    fuel_type[0:]
except NoSuchElementException:
    fuel_type.append('-')
time.sleep(1)

#extracting number of owners

try:
    for i in range(500):
        owner.append('1st')
    owner[0:]
except NoSuchElementException:
    owner.append('-')
time.sleep(1)

    
#extracting price
try:
    price_tags=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-5 text-right price']/span")
    for i in price_tags:
        prices=i.text.replace('₹','')
        p=prices.replace(',','')
        price.append(p)
    price[0:]
except NoSuchElementException:
    price.append('-')
time.sleep(1)



Pun2=pd.DataFrame({'Model Year': model_num[0:500],'Brand': brand[0:500],
                 'Car Name': name[0:500],'Variant': variant[0:500],'Distance Travelled': distance[0:500],
                 'Number of Owners': owner[0:500],'fuel type': fuel_type[0:500],'Price': price[0:500]})
```
